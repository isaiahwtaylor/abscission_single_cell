```python
%matplotlib inline
import scrublet as scr
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
wt_1_counts_matrix = scipy.io.mmread("/home/robotmessenger810/sc_analysis/data/raw_count_matrices/sc_counts_post_empty_drops/wt_1_matrix.mtx").T.tocsc()
wt_2_counts_matrix = scipy.io.mmread("/home/robotmessenger810/sc_analysis/data/raw_count_matrices/sc_counts_post_empty_drops/wt_2_matrix.mtx").T.tocsc()
mut_1_counts_matrix = scipy.io.mmread("/home/robotmessenger810/sc_analysis/data/raw_count_matrices/sc_counts_post_empty_drops/mut_1_matrix.mtx").T.tocsc()
mut_2_counts_matrix = scipy.io.mmread("/home/robotmessenger810/sc_analysis/data/raw_count_matrices/sc_counts_post_empty_drops/mut_2_matrix.mtx").T.tocsc()
```

```python
wt_1_scrub = scr.Scrublet(wt_1_counts_matrix, expected_doublet_rate=0.075, sim_doublet_ratio = 10)
wt_2_scrub = scr.Scrublet(wt_2_counts_matrix, expected_doublet_rate=0.075, sim_doublet_ratio = 10)
mut_1_scrub = scr.Scrublet(mut_1_counts_matrix, expected_doublet_rate=0.075, sim_doublet_ratio = 10)
mut_2_scrub = scr.Scrublet(mut_2_counts_matrix, expected_doublet_rate=0.075, sim_doublet_ratio = 10)
```

```python
wt_1_doublet_scores, wt_1_predicted_doublets = wt_1_scrub.scrub_doublets()
wt_2_doublet_scores, wt_2_predicted_doublets = wt_2_scrub.scrub_doublets()
mut_1_doublet_scores, mut_1_predicted_doublets = mut_1_scrub.scrub_doublets()
mut_2_doublet_scores, mut_2_predicted_doublets = mut_2_scrub.scrub_doublets()
```

```python
wt_1_scrub.plot_histogram()
wt_2_scrub.plot_histogram()
mut_1_scrub.plot_histogram()
mut_2_scrub.plot_histogram()
```

```python
print(sum(wt_1_doublet_scores > .2))
print(sum(wt_2_doublet_scores > .19))
print(sum(mut_1_doublet_scores > .23))
print(sum(mut_2_doublet_scores > .22))
```

```python
print(np.shape(wt_1_counts_matrix[wt_1_doublet_scores < .2,:]))
print(np.shape(wt_2_counts_matrix[wt_2_doublet_scores < .19,:]))
print(np.shape(mut_1_counts_matrix[mut_1_doublet_scores < .23,:]))
print(np.shape(mut_2_counts_matrix[mut_2_doublet_scores < .22,:]))
```
