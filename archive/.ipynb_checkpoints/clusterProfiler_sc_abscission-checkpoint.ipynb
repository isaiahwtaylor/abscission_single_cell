```r
library("dplyr")
library("clusterProfiler")
library("org.At.tair.db")
library("AnnotationHub")
library("PANTHER.db")
```

```r
cluster_AZ_all = readRDS(paste("../data/markers/", "AZ_markers_WT_ALL_res_1_25_Feb_14_22", ".rds", sep=""))
```

```r
head(cluster_AZ_all)
```

```r
cluster_AZ_all %>% group_by(cluster) %>% top_n(n = 2, wt = avg_logFC)
```

```r
dim(cluster_AZ_all)
```

```r
##################################################################
#Subsetting top 100 markers with adjusted p values lower than .05#
##################################################################
top100 <- cluster_AZ_all  %>% group_by(cluster) %>% top_n(n = 500, wt = avg_logFC)
top100pval <- subset(top100, rowSums(top100[5] < 0.05) > 0)
```

```r
df <- top100pval[,7:6]
dfsample <- split(df$gene,df$cluster)
length(dfsample)
```

```r
df[df$cluster == 9,]
```

```r
dfsample$`0` = bitr(dfsample$`0`, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
dfsample$`1` = bitr(dfsample$`1`, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
dfsample$`2` = bitr(dfsample$`2`, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
dfsample$`3` = bitr(dfsample$`3`, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
dfsample$`4` = bitr(dfsample$`4`, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
dfsample$`5` = bitr(dfsample$`5`, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
dfsample$`6` = bitr(dfsample$`6`, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
dfsample$`7` = bitr(dfsample$`7`, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
dfsample$`8` = bitr(dfsample$`8`, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
dfsample$`9` = bitr(dfsample$`9`, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
```

```r
AZ = bitr(cluster_AZ_all[cluster_AZ_all$cluster == 9 & cluster_AZ_all$avg_logFC > 2 & cluster_AZ_all$p_val_adj < .1,]$gene, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
universe = bitr(cluster_AZ_all[cluster_AZ_all$cluster == 9,]$gene, fromType="TAIR", toType="ENTREZID", OrgDb="org.At.tair.db")
```

```r
dim(universe)
dim(AZ)
```
