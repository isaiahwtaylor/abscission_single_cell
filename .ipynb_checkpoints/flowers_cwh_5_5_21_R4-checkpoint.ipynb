```r
flower_seu <- FindClusters(flower_seu, resolution = 0.3)
DimPlot(flower_seu, reduction = "umap", label = FALSE, pt.size = 1.5)
```

```r
(ident_vec = unique(Idents(flower_seu)))
```

```r
#get pseudobulk for each cluster to compare with kwak data
flower_seu@active.assay = "RNA"
pbs = data.frame(matrix(nrow = dim(flower_seu@assays$RNA@counts)[1], ncol =length(ident_vec)))
count = 1
for (l in ident_vec) {
    pbs[,count] =  rowSums(as.matrix(GetAssayData(subset(flower_seu, ident = l), slot = "counts")))
#    pbs[[count]] = rowSums(as.matrix(GetAssayData(subset(flower_seu), subset = orig.geno == "WT"), slot = "counts")[, WhichCells(subset(flower_seu, subset = orig.geno == "WT"), ident = l)])) 
    count = count + 1
}

colnames(pbs) = ident_vec
rownames(pbs) = rownames(as.matrix(GetAssayData(subset(flower_seu, ident = l), slot = "counts")))

pbs_cpm = t(t(pbs)/colSums(pbs) * 1000000)
#write.csv(pbs_cpm, here("../data", "pbs_cpm_5_5_21.csv"))
```
