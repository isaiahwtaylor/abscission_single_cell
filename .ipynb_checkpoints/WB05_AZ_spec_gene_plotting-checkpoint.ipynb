```r
resolution = 1.25
set.seed(42)
DefaultAssay(seu_intd_wt_orig) <- "integrated"
options(repr.plot.width=12, repr.plot.height=12)
# Run the standard workflow for visualization and clustering
#all_intd_sct <- ScaleData(all_intd_sct, verbose = FALSE)
seu_intd_wt_orig <- RunPCA(seu_intd_wt_orig, npcs = 100, verbose = FALSE, approx = FALSE)
#From RunPCA doc: Features to compute PCA on. If features=NULL, PCA will be run using the variable features for the Assay. 
#Note that the features must be present in the scaled data. Any requested features that are not scaled or have 0 variance 
#will be dropped, and the PCA will be run using the remaining features.

#previously run 20 PCs as of 2/14/22
seu_intd_wt_orig <- FindNeighbors(seu_intd_wt_orig, dims = 1:20, verbose = FALSE)
seu_intd_wt_orig <- FindClusters(seu_intd_wt_orig, resolution = resolution, algorithm = 3, verbose = FALSE)
seu_intd_wt_orig <- RunUMAP(seu_intd_wt_orig, reduction = "pca", dims = 1:20, verbose = FALSE)
```

```r
trava_counts = counts_to_reads_df("/home/robotmessenger810/sc_analysis/data/bulk_data/trava")
```

```r
colnames(trava_counts) = c("car_1","car_2","fil_1","fil_2","pet_1","pet_2","sep_1","sep_2","sil_1","sil_2")
```

```r
#make DGElist
x3 <- DGEList(counts = trava_counts, genes = rownames(trava_counts))


#reads per library uniquely mapped to a gene:

#make cpm and lcpm
cpm <- cpm(x3)
lcpm <- cpm(x3, log=TRUE)

#keep only genes that are expressed. "Expressed" here means counts observed in at least 3 samples
dim(x3)
keep.exprs <- rowMeans(cpm)>=.5
x3 <- x3[keep.exprs,, keep.lib.sizes=FALSE]
dim(x3) #compare to dim(x) above

#normalize data after removing low expressed genes
x3 <- calcNormFactors(x3)

#cpm, lcpm of normalized values
cpm <- cpm(x3)
lcpm <- cpm(x3, log=TRUE)

```

```r
cpm["AT4G23600",]
```

```r
fil_top = data.frame(Reduce("/", list(rowMeans(cpm[,c(3,4)]), rowMeans(.1+cpm[,c(5:10)]))))
rownames(fil_top) = rownames(cpm)
rs =  rownames(cpm)[order(fil_top, decreasing = TRUE)]
fil_top = data.frame(fil_top[order(fil_top, decreasing = TRUE),])
rownames(fil_top) = rs

pet_top = data.frame(Reduce("/", list(rowMeans(cpm[,c(5,6)]), rowMeans(.1+cpm[,c(3,4,7,8,9,10)]))))
rownames(pet_top) = rownames(cpm)
rs =  rownames(cpm)[order(pet_top, decreasing = TRUE)]
pet_top = data.frame(pet_top[order(pet_top, decreasing = TRUE),])
rownames(pet_top) = rs

sep_top = data.frame(Reduce("/", list(rowMeans(cpm[,c(7,8)]), rowMeans(.1+cpm[,c(3,4,5,6,9,10)]))))
rownames(sep_top) = rownames(cpm)
rs =  rownames(cpm)[order(sep_top, decreasing = TRUE)]
sep_top = data.frame(sep_top[order(sep_top, decreasing = TRUE),])
rownames(sep_top) = rs

sil_top = data.frame(Reduce("/", list(rowMeans(cpm[,c(9,10)]), rowMeans(.1+cpm[,c(3,4,5,6,7,8)]))))
rownames(sil_top) = rownames(cpm)
rs =  rownames(cpm)[order(sil_top, decreasing = TRUE)]
sil_top = data.frame(sil_top[order(sil_top, decreasing = TRUE),])
rownames(sil_top) = rs
```
